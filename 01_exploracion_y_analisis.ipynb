```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
import numpy as np

# Features (X) y etiquetas (y)
X = data.drop(data.columns[-3:], axis=1)  # Features, excluyendo las últimas 3 columnas
y = data.iloc[:, -3:]  # Últimas 3 columnas, que son las etiquetas

# Dividir los datos en entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')  # También puedes usar 'median' o 'most_frequent'

# Ajustar y transformar el conjunto de entrenamiento
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)

# Ajustar y transformar el conjunto de rest
X_test = pd.DataFrame(imputer.transform(X_test), columns=X_train.columns)
```

```python
from scipy.stats import shapiro, kstest

for column in X_train.columns:
    data_cont = X_train[column]
    
    # Prueba de Shapiro-Wilk
    stat_shapiro, p_value_shapiro = shapiro(data_cont)
    
    # Prueba de Kolmogorov-Smirnov
    stat_ks, p_value_ks = kstest(data_cont, 'norm', args=(np.mean(data_cont), np.std(data_cont)))

    print(f'Columna: {column}')
    print(f'  Shapiro-Wilk p-value: {p_value_shapiro}')
    print(f'  Kolmogorov-Smirnov p-value: {p_value_ks}')
    
    if p_value_shapiro > 0.05 and p_value_ks > 0.05:
        print(f"  -> Los datos en {column} parecen provenir de una distribución normal.")
    else:
        print(f"  -> Los datos en {column} no parecen provenir de una distribución normal.")
    print()
```

```python
# Lista para almacenar resultados de cada iteración
results = []

# Realizar 50 iteraciones variando el random_state (de 0 a 49)
for i in range(50):
    # 1. Dividir en train y test (15% para test)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=i)
    
    # 2. Dividir el conjunto train en train_2 y validation (15% del train para validación)
    X_train_2, X_val, y_train_2, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=i)
    
    # 3. Combinar train_2 y validación para la búsqueda de hiperparámetros
    X_combined = pd.concat([X_train_2, X_val], axis=0)
    y_combined = pd.concat([y_train_2, y_val], axis=0)
    
    # 4. Definir el modelo base RandomForestRegressor
    model = RandomForestRegressor(random_state=42)
    
    # 5. Definir la grilla de hiperparámetros
    param_grid = {
        'n_estimators': [15, 20, 30, 50],
        'max_depth': [None, 5, 10, 15],
        'min_samples_split': [1, 2, 5],
        'min_samples_leaf': [1, 2, 5],
        'max_features': ['auto', 'sqrt']
    }
    
    # 6. Configurar HalvingGridSearchCV
    halving_cv = HalvingGridSearchCV(
        estimator=model,
        param_grid=param_grid,
        factor=2,                         # Descarta la mitad de las combinaciones en cada iteración
        resource='n_samples',             # Utiliza el número de muestras como recurso
        max_resources='auto',             # Se adapta automáticamente al tamaño del conjunto
        scoring='neg_mean_squared_error', # Métrica para minimizar (MSE negativo)
        cv=3,                             # Validación cruzada interna de 3 pliegues
        verbose=0,
        n_jobs=-1,                         # Usa todos los núcleos disponibles
        error_score=np.nan
    )
    
    # 7. Ejecutar la búsqueda de hiperparámetros sobre el conjunto combinado
    halving_cv.fit(X_combined, y_combined)
    
    # 8. Extraer los mejores hiperparámetros y evaluar en el conjunto de test
    best_params = halving_cv.best_params_
    test_score = halving_cv.score(X_test, y_test)
    
    # 9. Almacenar los resultados de la iteración
    results.append({
        'random_state': i,
        'best_params': best_params,
        'test_score': test_score
    })
    
# Convertir la lista de resultados a un DataFrame para análisis
df_results = pd.DataFrame(results)
print("Resultados de las primeras iteraciones:")
print(df_results.head())

# Calcular la puntuación promedio en el conjunto de test a lo largo de las 50 iteraciones
average_score = df_results['test_score'].mean()
print("\nPuntuación promedio en test:", average_score)

# Extraer y combinar los hiperparámetros para obtener un resumen estadístico
df_params = pd.json_normalize(df_results['best_params'])
df_combined = pd.concat([df_results.drop(columns=['best_params']), df_params], axis=1)
print("\nResumen estadístico de los hiperparámetros:")
print(df_combined.describe())

# IDENTIFICAR LA MEJOR ITERACIÓN:
# Como se usa 'neg_mean_squared_error', cuanto mayor sea el valor (más cercano a 0), mejor es el desempeño.
best_iteration = df_results.loc[df_results['test_score'].idxmax()]
print("\nLa mejor iteración es:")
print(best_iteration['best_params'])
```

```python
# Lista para almacenar resultados de cada iteración
results = []

# Realizar 50 iteraciones variando el random_state (de 0 a 49)
for i in range(50):
    # 1. Dividir en train y test (15% para test)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=i)
    
    # 2. Dividir el conjunto train en train_2 y validación (15% del train para validación)
    X_train_2, X_val, y_train_2, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=i)
    
    # 3. Combinar train_2 y validación para la búsqueda de hiperparámetros
    X_combined = pd.concat([X_train_2, X_val], axis=0)
    y_combined = pd.concat([y_train_2, y_val], axis=0)
    
    # 4. Definir el modelo base DecisionTreeRegressor
    model = DecisionTreeRegressor(random_state=42)
    
    # 5. Definir la grilla de hiperparámetros
    param_grid = {
        'max_depth': [3, 4, 5, 7],
        'min_samples_split': [2, 5, 30],
        'min_samples_leaf': [1, 2, 4],
        'max_features': [None, 'sqrt', 'log2']
    }
    
    # 6. Configurar HalvingGridSearchCV
    halving_cv = HalvingGridSearchCV(
        estimator=model,
        param_grid=param_grid,
        factor=2,                         # Reduce a la mitad el número de combinaciones en cada iteración
        resource='n_samples',             # Utiliza el número de muestras como recurso
        max_resources='auto',             # Se adapta automáticamente al tamaño del conjunto
        scoring='neg_mean_squared_error', # Métrica para minimizar (MSE negativo)
        cv=3,                             # Validación cruzada interna de 3 pliegues
        verbose=0,
        n_jobs=-1,                         # Utiliza todos los núcleos disponibles
        error_score=np.nan
    )
    
    # 7. Ejecutar la búsqueda de hiperparámetros en el conjunto combinado
    halving_cv.fit(X_combined.values, y_combined.values)
    
    # 8. Extraer los mejores hiperparámetros y evaluar en el conjunto de test
    best_params = halving_cv.best_params_
    test_score = halving_cv.score(X_test.values, y_test.values)
    
    # 9. Almacenar los resultados de la iteración
    results.append({
        'random_state': i,
        'best_params': best_params,
        'test_score': test_score
    })
    
# Convertir la lista de resultados a un DataFrame para análisis
df_results = pd.DataFrame(results)
print("Resultados de las primeras iteraciones:")
print(df_results.head())

# Calcular la puntuación promedio en el conjunto de test a lo largo de las 50 iteraciones
average_score = df_results['test_score'].mean()
print("\nPuntuación promedio en test:", average_score)

# Extraer y combinar los hiperparámetros para obtener un resumen estadístico
df_params = pd.json_normalize(df_results['best_params'])
df_combined_results = pd.concat([df_results.drop(columns=['best_params']), df_params], axis=1)
print("\nResumen estadístico de los hiperparámetros:")
print(df_combined_results.describe())

# IDENTIFICAR LA MEJOR ITERACIÓN:
# Como se usa 'neg_mean_squared_error', cuanto mayor sea el valor (más cercano a 0), mejor es el desempeño.
best_iteration = df_results.loc[df_results['test_score'].idxmax()]
print("\nLa mejor iteración es:")
print(best_iteration['best_params'])
```

```python
# Lista para almacenar resultados de cada iteración
results = []

# Realizar 50 iteraciones variando el random_state (de 0 a 49)
for i in range(50):
    # 1. Dividir en train y test (15% para test)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=i)
    
    # 2. Dividir el conjunto train en train_2 y validación (15% del train para validación)
    X_train_2, X_val, y_train_2, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=i)
    
    # 3. Combinar train_2 y validación para la búsqueda de hiperparámetros
    X_combined = pd.concat([X_train_2, X_val], axis=0)
    y_combined = pd.concat([y_train_2, y_val], axis=0)
    
    # 4. Definir el modelo base CatBoostRegressor (se desactiva la salida con verbose=0)
    cat_model = CatBoostRegressor(verbose=0, random_state=42)
    # Envolver el modelo para manejar multi-target
    cat_model_multi = MultiOutputRegressor(cat_model)
    
    # 5. Definir la grilla de hiperparámetros (usar el prefijo 'estimator__' para acceder al modelo interno)
    param_grid = {
        'estimator__iterations': [100, 200, 300, 500],
        'estimator__depth': [3, 4, 5],
        'estimator__learning_rate': [0.1, 0.01, 0.001, 0.0001],
        'estimator__l2_leaf_reg': [1, 3, 5, 7]
    }
    
    # 6. Configurar HalvingGridSearchCV
    halving_cv = HalvingGridSearchCV(
        estimator=cat_model_multi,
        param_grid=param_grid,
        factor=2,                         # Descarta la mitad de las combinaciones en cada iteración
        resource='n_samples',             # Usa el número de muestras como recurso
        max_resources='auto',             # Se adapta automáticamente al tamaño del conjunto
        scoring='neg_mean_squared_error', # Métrica a minimizar (MSE negativo)
        cv=3,                             # Validación cruzada interna de 3 pliegues
        verbose=0,
        n_jobs=-1,                         # Utiliza todos los núcleos disponibles
        error_score=np.nan
    )
    
    # 7. Ejecutar la búsqueda de hiperparámetros sobre el conjunto combinado
    # Convertimos los DataFrames a arrays para evitar warnings de nombres de columnas
    halving_cv.fit(X_combined.values, y_combined.values)
    
    # 8. Extraer los mejores hiperparámetros y evaluar en el conjunto de test
    best_params = halving_cv.best_params_
    test_score = halving_cv.score(X_test.values, y_test.values)
    
    # 9. Almacenar los resultados de la iteración
    results.append({
        'random_state': i,
        'best_params': best_params,
        'test_score': test_score
    })
    
# Convertir la lista de resultados a un DataFrame para análisis
df_results = pd.DataFrame(results)
print("Resultados de las primeras iteraciones:")
print(df_results.head())

# Calcular la puntuación promedio en el conjunto de test a lo largo de las 50 iteraciones
average_score = df_results['test_score'].mean()
print("\nPuntuación promedio en test:", average_score)

# Extraer y combinar los hiperparámetros para obtener un resumen estadístico
df_params = pd.json_normalize(df_results['best_params'])
df_combined_results = pd.concat([df_results.drop(columns=['best_params']), df_params], axis=1)
print("\nResumen estadístico de los hiperparámetros:")
print(df_combined_results.describe())

# IDENTIFICAR LA MEJOR ITERACIÓN:
# Como se usa 'neg_mean_squared_error', cuanto mayor sea el valor (más cercano a 0), mejor es el desempeño.
best_iteration = df_results.loc[df_results['test_score'].idxmax()]
print("\nLa mejor iteración es:")
print(best_iteration['best_params'])
```

```python
# Lista para almacenar resultados de cada iteración
results = []

# Realizar 50 iteraciones variando el random_state (de 0 a 49)
for i in range(50):
    # 1. Dividir en train y test (15% para test)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=i)
    
    # 2. Dividir el conjunto train en train_2 y validación (15% del train para validación)
    X_train_2, X_val, y_train_2, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=i)
    
    # 3. Combinar train_2 y validación para la búsqueda de hiperparámetros
    X_combined = pd.concat([X_train_2, X_val], axis=0)
    y_combined = pd.concat([y_train_2, y_val], axis=0)
    
    # 4. Definir el modelo base LGBMRegressor y envolverlo para multi-target
    base_model = LGBMRegressor(random_state=42)
    model = MultiOutputRegressor(base_model)
    
    # 5. Definir la grilla de hiperparámetros
    # Usamos el prefijo 'estimator__' para acceder a los parámetros del estimador interno
    param_grid = {
        'estimator__n_estimators': [150, 200, 250],
        'estimator__max_depth': [4, 7, 10],
        'estimator__learning_rate': [0.01, 0.001, 0.0001],
        'estimator__num_leaves': [40, 50, 60]
    }
    
    # 6. Configurar HalvingGridSearchCV
    halving_cv = HalvingGridSearchCV(
        estimator=model,
        param_grid=param_grid,
        factor=2,                         # Descarta la mitad de las combinaciones en cada iteración
        resource='n_samples',             # Utiliza el número de muestras como recurso
        max_resources='auto',             # Se adapta automáticamente al tamaño del conjunto
        scoring='neg_mean_squared_error', # Métrica para minimizar (MSE negativo)
        cv=3,                             # Validación cruzada interna de 3 pliegues
        verbose=0,
        n_jobs=-1,                         # Usa todos los núcleos disponibles
        error_score=np.nan
    )
    
    # 7. Ejecutar la búsqueda de hiperparámetros sobre el conjunto combinado
    # Convertimos a arrays para evitar warnings de nombres de columnas
    halving_cv.fit(X_combined.values, y_combined.values)
    
    # 8. Extraer los mejores hiperparámetros y evaluar en el conjunto de test
    best_params = halving_cv.best_params_
    test_score = halving_cv.score(X_test.values, y_test.values)
    
    # 9. Almacenar los resultados de la iteración
    results.append({
        'random_state': i,
        'best_params': best_params,
        'test_score': test_score
    })
    
# Convertir la lista de resultados a un DataFrame para análisis
df_results = pd.DataFrame(results)
print("Resultados de las primeras iteraciones:")
print(df_results.head())

# Calcular la puntuación promedio en el conjunto de test a lo largo de las 50 iteraciones
average_score = df_results['test_score'].mean()
print("\nPuntuación promedio en test:", average_score)

# Extraer y combinar los hiperparámetros para obtener un resumen estadístico
df_params = pd.json_normalize(df_results['best_params'])
df_combined_results = pd.concat([df_results.drop(columns=['best_params']), df_params], axis=1)
print("\nResumen estadístico de los hiperparámetros:")
print(df_combined_results.describe())

# IDENTIFICAR LA MEJOR ITERACIÓN:
# Como se usa 'neg_mean_squared_error', cuanto mayor sea el valor (más cercano a 0), mejor es el desempeño.
best_iteration = df_results.loc[df_results['test_score'].idxmax()]
print("\nLa mejor iteración es:")
print(best_iteration['best_params'])
```
